```python
import itertools
dates=[]
rain=[]
for key, value in dataSM.iteritems():
    ayear=int(key)
    sdate=date(ayear,1,1)
    edate=date(ayear+1,1,1)
    dates.append(pd.date_range(sdate,edate-timedelta(days=1),freq='d'))
    rain.append(dataSM[key])
    #print(key,"\t",ayear+1)
flattened_dates = itertools.chain.from_iterable(dates)
flattened_rain = itertools.chain.from_iterable(rain)
dailySMrain=pd.DataFrame([flattened_dates,flattened_rain]).T
dailySMrain.columns=["Date","Rain[mm]"]
dailySMrain=dailySMrain.set_index("Date")
dailySMrain
```
