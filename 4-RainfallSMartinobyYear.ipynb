```python
dSm=pd.read_feather('DailyRainfallSMartino.fea')
dSm=dSm.set_index("Date")
```

```python
dSm.plot(figsize=(20,4))
```

```python
annual_precipitation_SM=dSm.resample("1YS").sum()
```

```python
annual_precipitation_SM.columns=["Annual Precipitation [mm]"]
annual_precipitation_SM
```

```python
annual_precipitation_SM.plot.bar(figsize=(20,4))
```

Question: how to put just the years in the abscissa and not the whole date ?

```python
annual_precipitation_SM.describe()
```

```python
c_prc_SM=annual_precipitation_SM.cumsum()
c_prc_SM.columns=["Cumulated Precipitation [mm]"]
c_prc_SM.plot(figsize=(20,4))
```
