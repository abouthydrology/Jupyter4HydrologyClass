```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
os.listdir()
```

```python
data = pd.ExcelFile('DatiEstremiPluviometriciFeltre.xlsx')
```

```python
data
```

```python
data.sheet_names
```

```python
df = data.parse('Sheet 1',header=2)
df
```

```python
df.columns
```

```python
df.rename(columns={'Unnamed: 0':'Anno'},inplace=True)
df.set_index("Anno",inplace=True)
df
```

```python
ax=df.plot(figsize=(15,15))
ax.set_lyabel="Precipitation [mm]"
ax
```
